# Publication Ready Figures

Below are some examples for making publication ready figures. You can include the figures in your LaTeX document without rescaling. For example, all you need to do is:

```
\begin{figure}[h]
\centering
\includegraphics{Figures/HalfTextWidth.pdf}
\caption{A generic half-width figure.}
\label{fig:halfwidth}
\end{figure}
```

```
\begin{figure}[h]
\centering
\includegraphics{Figures/FullTextWidth.pdf}
\caption{A generic full-width figure.}
\label{fig:fullwidth}
\end{figure}
```


````python
%pylab inline
%config InlineBackend.figure_format='retina' 
# from IPython.display import set_matplotlib_formats
# set_matplotlib_formats('png','pdf')
````

````python
x = np.linspace(0, 10, 100)
y = x ** 2
````

## Default fonts

````python
# save the inline defaults
# inline_rc = dict(rcParams)
# matplotlib.rcParams.update(matplotlib.rcParamsDefault)

# use: mpl.rcParams.update(inline_rc) to restore defults

# set font to 9pt
matplotlib.rcParams.update({'font.size': 9})

# ------ Half Page Width --------------------

width = 3.
height = 2.33

fig, ax = plt.subplots(figsize=(width,height))

ax.plot(x, x**2, label=r"$y = x^2$")
ax.plot(x, x**3, label=r"$y = x^3$")
ax.plot(x, -x**2.5, label=r"$y = -x^{2.5}$")
ax.set_xlabel(r'Text $x$ [-]')
ax.set_ylabel(r'Text $y$ [-]')
ax.legend()
savefig('HalfTextWidth.pdf',bbox_inches="tight", pad_inches=0.005);

# ------ Full Page Width --------------------

width = 6
height = 3

fig, ax = plt.subplots(figsize=(width,height))

ax.plot(x, x**2, label=r"$y = x^2$")
ax.plot(x, x**3, label=r"$y = x^3$")
ax.plot(x, -x**2.5, label=r"$y = -x^{2.5}$")
ax.set_xlabel(r'Text $x$ [-]')
ax.set_ylabel(r'Text $y$ [-]')
ax.legend()
savefig('FullTextWidth.pdf',bbox_inches="tight", pad_inches=0.005);
````

## Change the color order

````python
matplotlib.rcParams.update({'font.size': 9})

# ------ Half Page Width --------------------

width = 3.
height = 2.33

fig, ax = plt.subplots(figsize=(width,height))

ax.plot(x, x**2, color='C3', label=r"$y = x^2$")
ax.plot(x, x**3, color='C0', label=r"$y = x^3$")
ax.plot(x, -x**2.5,color='C2', label=r"$y = -x^{2.5}$")
ax.set_xlabel(r'Text $x$ [-]')
ax.set_ylabel(r'Text $y$ [-]')
ax.legend()
savefig('HalfTextWidth.pdf',bbox_inches="tight", pad_inches=0.005);

# ------ Full Page Width --------------------

width = 6
height = 3

fig, ax = plt.subplots(figsize=(width,height))

ax.plot(x, x**2, color='C3', label=r"$y = x^2$")
ax.plot(x, x**3, color='C0', label=r"$y = x^3$")
ax.plot(x, -x**2.5,color='C2', label=r"$y = -x^{2.5}$")
ax.set_xlabel(r'Text $x$ [-]')
ax.set_ylabel(r'Text $y$ [-]')
ax.legend()
savefig('FullTextWidth.pdf',bbox_inches="tight", pad_inches=0.005);
````
